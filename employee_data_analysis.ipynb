```python
# Task 2
# The cost column in the dataframe “Project” has some missing values. 
# Your task is to compute these missing values.
# Replace the missing values by running average. 
# You should use the “For” loop for this task.
```

```python
# reading the file
project_df = pd.read_csv("project_data.csv")

# checking data before filling
print("Before filling missing values:\n")
print(project_df)

# finding how many missing values are there
print("Missing values:", project_df["Cost"].isna().sum())

# variables to keep total and count
total = 0
count = 0

# copying the cost column
costs = project_df["Cost"].values.copy()

# loop to fill missing values
for i in range(len(costs)):
    if np.isnan(costs[i]):
        if count > 0:
            avg = total / count
        else:
            avg = 0
        costs[i] = avg
        print("Filled value at index", i, "with", avg)
    total += costs[i]
    count += 1

# updating the dataframe
project_df["Cost"] = costs

# printing after filling
print("\nAfter filling missing values:\n")
print(project_df)
print("Missing values:", project_df["Cost"].isna().sum())

# saving the file
project_df.to_csv("project_data.csv", index=False)
print("\nFile saved.")
```

```python
# Task 3
# Split the name column in the Employee dataframe into two new columns.
# “First Name”, and “Last Name” and remove the older “name” column.
```

```python
employee_df = pd.read_csv("employee_data.csv") 

print("Before splitting Name column:\n")
print(employee_df)
print("\nColumns:", list(employee_df.columns))

# Splitting the Name column into two parts: First Name and Last Name
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', n=1, expand=True)

# Droping the old Name column
employee_df = employee_df.drop('Name', axis=1)

# Reordering columns for better readability
column_order = ['ID', 'First Name', 'Last Name', 'Gender', 'City', 'Age']
employee_df = employee_df[column_order]

print("\n" + "="*60)
print("After splitting Name column:")
print("="*60)
print(employee_df)
print("\nColumns:", list(employee_df.columns))

# then saving the cleaned file
employee_df.to_csv("employee_data.csv", index=False)
print("\n File saved successfully")
```

```python
# Task 4
# Join all three dataframes in one single dataframe. Name it “Final”
```

```python
# reading all csv files
employee_df = pd.read_csv("employee_data.csv")
seniority_df = pd.read_csv("seniority_data.csv")
project_df = pd.read_csv("project_data.csv")

# checking if files are loaded
print("Employee data:\n", employee_df)
print("\nSeniority data:\n", seniority_df)
print("\nProject data:\n", project_df)

# merging employee and seniority data
temp = pd.merge(employee_df, seniority_df, on='ID', how='inner')

# merging the result with project data
Final = pd.merge(temp, project_df, on='ID', how='inner')

# printing final merged data
print("\nFinal merged data:\n", Final)

# saving the file
Final.to_csv("final_data.csv", index=False)
print("\nFile saved as final_data.csv")

# checking some details
print("\nTotal rows:", len(Final))
print("Unique employees:", Final['ID'].nunique())
print("\nProjects per employee:")
print(Final['ID'].value_counts())
```

```python
# Task 5
# Add a new bonus column in the Final dataframe.
# Give a 5% bonus concerning project cost only to employees who have finished the projects.
```

```python
# reading the data
Final = pd.read_csv("final_data.csv")

# checking before adding bonus
print("Before adding Bonus column:\n")
print(Final)

# adding bonus column
Final['Bonus'] = 0
for i in range(len(Final)):
    if Final.loc[i, 'Status'] == 'Finished':
        Final.loc[i, 'Bonus'] = Final.loc[i, 'Cost'] * 0.05

# printing after adding bonus
print("\nAfter adding Bonus column:\n")
print(Final)

# checking total bonus
print("\nTotal Bonus amount:", Final['Bonus'].sum())
print("Projects with bonus:", (Final['Bonus'] > 0).sum())
print("Projects without bonus:", (Final['Bonus'] == 0).sum())

# saving the file
Final.to_csv("final_data.csv", index=False)
print("\nFile saved successfully.")
```

```python
 # Task 6
# Demote the designation level by 1, whose projects have status “fail”. 
# Delete the employees record whose designation level is above 4.
```

```python
# reading the data
Final = pd.read_csv("final_data.csv")

# checking before demotion
print("Before demotion:\n")
print(Final[['ID','First Name','Last Name','Designation Level','Project','Status']])

# counting failed projects and demoting
for i in range(len(Final)):
    if Final.loc[i, 'Status'] == 'Failed':
        Final.loc[i, 'Designation Level'] = Final.loc[i, 'Designation Level'] + 1

# printing after demotion
print("\nAfter demotion:\n")
print(Final[['ID','First Name','Last Name','Designation Level','Project','Status']])

# removing employees with level > 4
Final = Final[Final['Designation Level'] <= 4]

# printing cleaned data
print("\nAfter removing disqualified employees:\n")
print(Final[['ID','First Name','Last Name','Designation Level','Project','Status']])

# saving file
Final.to_csv("final_data.csv", index=False)
print("\nFile updated and saved.")
```

```python
# Task 7
# Add “Mr.” and “Mrs.” to the first name column and drop the gender column.
```

```python
# reading the data
Final = pd.read_csv("final_data.csv")

# checking before change
print("Before adding titles:\n")
print(Final[['ID','First Name','Last Name','Gender']])

# adding titles using a simple for loop
for i in range(len(Final)):
    if Final.loc[i, 'Gender'] == 'M':
        Final.loc[i, 'First Name'] = "Mr. " + Final.loc[i, 'First Name']
    else:
        Final.loc[i, 'First Name'] = "Mrs. " + Final.loc[i, 'First Name']

# printing after adding titles
print("\nAfter adding titles:\n")
print(Final[['ID','First Name','Last Name','Gender']])

# dropping gender column
Final = Final.drop('Gender', axis=1)

# checking final dataframe
print("\nAfter dropping Gender column:\n")
print(Final)

# saving file
Final.to_csv("final_data.csv", index=False)
print("\nFile updated and saved.")
```

```python
# Task 8
# Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.
```

```python
# reading the data
Final = pd.read_csv("final_data.csv")
# checking data before promotion
print("Before promotion:\n")
print(Final[['ID','First Name','Last Name','Age','Designation Level']])

# loop through each row
for i in range(len(Final)):
    age = Final.loc[i, 'Age']
    level = Final.loc[i, 'Designation Level']
    # if age is more than 29, decrease level by 1
    if age > 29:
        if level > 1:
            Final.loc[i, 'Designation Level'] = level - 1
        else:
            Final.loc[i, 'Designation Level'] = 1

# printing after promotion
print("\nAfter promotion:\n")
print(Final[['ID','First Name','Last Name','Age','Designation Level']])

# saving the file
Final.to_csv("final_data.csv", index=False)
print("\nFile updated and saved.")
```

```python
# Task 9
# Add the cost of all projects for each Employee and save it in new dataframe
# “TotalProjCost” with three columns ID, First Name, and Total cost
```

```python
# reading the data
Final = pd.read_csv("final_data.csv")

# checking data
print("Project data:\n")
print(Final[['ID','First Name','Project','Cost']])

# calculating total cost for each employee
TotalProjCost = Final.groupby(['ID','First Name'])['Cost'].sum().reset_index()

# renaming the column
TotalProjCost.rename(columns={'Cost':'Total Cost'}, inplace=True)

# printing result
print("\nTotal project cost per employee:\n")
print(TotalProjCost)

# saving the new file
TotalProjCost.to_csv("TotalProjCost.csv", index=False)
print("\nFile saved as TotalProjCost.csv")
```

```python
import pandas as pd

# reading the data
Final = pd.read_csv("final_data.csv")

# checking city data
print("Employee and City data:\n")
print(Final[['ID','First Name','Last Name','City']])

# filtering cities that have letter 'o'
filtered = Final[Final['City'].str.contains('o', case=False, na=False)]

# printing result
print("\nEmployees whose city name has letter 'o':\n")
print(filtered[['ID','First Name','Last Name','City']])

# saving filtered data
filtered.to_csv("Employees_with_o.csv", index=False)
print("\nFile saved as Employees_with_o.csv")
```
